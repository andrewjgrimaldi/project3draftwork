```python
#to import regular expression tags 
import re 
```

```python
dollar_amounts = re.findall(r"\$\d+(?:\,\d+)?", results2[0].text)
dollar_amounts
```

```python
#list comprehension (instead of for loop) to iterate over and grab the max
numeric_dollar_amounts = [int(re.sub(",","",x[1:])) for x in dollar_amounts]
print(numeric_dollar_amounts)
```

```python
#get max
principal_amount = max(numeric_dollar_amounts)
```

```python
#look for keyword
# #first try
# "deposit of ten percent"

# else

# #regular expression match
# then:
#     "deposit of $" 
```

```python
deposit = re.findall(r"deposit of \$\d+(?:\,\d+)?", results2[0].text.lower())
deposit2 = [int(re.sub(",","",x[12:])) for x in deposit]
deposit2
```

```python

nodeposit = "deposit of ten percent" in results2[0].text.lower()
nodeposit
```

```python
def get_deposit(principal_amount, results2):

    if nodeposit:
        return 0.1 * principal_amount
    else:
        deposit = re.findall(r"deposit of \$\d+(?:\,\d+)?", results2[0].text.lower())
        deposit2 = [int(re.sub(",","",x[12:])) for x in deposit]
        return deposit2[0]

```

```python
get_deposit(principal_amount, results2)
```

```python
#to get address

property_address = re.findall(r"trustee sale.*\d{5}", str(results2[0]))
property_address
```

```python
def get_address(results2):

    a = results2[0]
    b = a.find_all("span")

    iter_b = iter(b)

    for i in range(len(b)):
        next_b = next(iter_b)
        if 'trustee sale' == next_b.text.lower():
            result = next(iter_b)
            result = result.text
            break

    return result
    
    
get_address(results2)
# for span in b:
#     print("trustee sale" in span.lower())
```
